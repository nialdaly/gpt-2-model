```python
import numpy as np
import pandas as pd
import missingno as msno
import re
import xlrd

%matplotlib inline
```

```python
metadata = pd.read_excel('../data/paper_metadata.xlsx', index_col=False)
metadata.head(3)
```

### data cleansing

```python
metadata = metadata.fillna(np.nan)

# removes trailing whitespace from metric names
metadata['metric_name'] = metadata['metric_name'].str.strip()
```

### missing data breakdown

```python
msno.matrix(metadata)
```

```python
# model sentences
model_sentences = []

# metrics
metric_name_sentences = []
metric_value_sentences = []

# datasets
dataset_sentences = []

# hyper-parameters
lr_sentences = []
bs_sentences = []
mom_sentences = []
epoch_sentences = []

# frameworks
framework_sentences = []
```

```python
for i in range(len(metadata)):
    
    model_sentences.append("The model described here is called " + str(metadata['model_name'].iloc[i]) + ". ")
    
    metric_name_sentences.append("The metric used to validate the model was " + str(metadata['metric_name'].iloc[i]) + " ")

    dataset_sentences.append("with the model being tested on the " + str(metadata['dataset'].iloc[i]) + " dataset. ")
    
    metric_value_sentences.append("A " + str(metadata['metric_name'].iloc[i]) + " of " + str(metadata['metric_value'].iloc[i]) + " was achieved. ")
    
    lr_sentences.append("The learning rate was set to " + str(metadata['learning_rate'].iloc[i]) + ". ")
    bs_sentences.append("A batch size of " + str(metadata['batch_size'].iloc[i]) + " was used. ")
    mom_sentences.append("A momentum value of " + str(metadata['momentum'].iloc[i]) + " was used. ")
    epoch_sentences.append("The model was trained with a total of " + str(metadata['epochs'].iloc[i]) + " epochs. ")
    
    framework_sentences.append("The " + str(metadata['framework'].iloc[i]) + " framework was used to develop the model. ")
```

### NaN sentence replacement

```python
# models
model_sentences = [x.replace(x, '') if 'nan' in str(x) else x for x in model_sentences]

# metrics
metric_name_sentences = [x.replace(x, '') if 'nan' in str(x) else x for x in metric_name_sentences]
metric_value_sentences = [x.replace(x, '') if 'nan' in str(x) else x for x in metric_value_sentences]

# datasets
dataset_sentences = [x.replace(x, '') if 'nan' in str(x) else x for x in dataset_sentences]

# hyper-parameters
lr_sentences = [x.replace(x, '') if 'nan' in str(x) else x for x in lr_sentences]
bs_sentences = [x.replace(x, '') if 'nan' in str(x) else x for x in bs_sentences]
mom_sentences = [x.replace(x, '') if 'nan' in str(x) else x for x in mom_sentences]
epoch_sentences = [x.replace(x, '') if 'nan' in str(x) else x for x in epoch_sentences]

# frameworks
framework_sentences = [x.replace(x, '') if 'nan' in str(x) else x for x in framework_sentences]
```

### data object generation

```python
abstracts = {}

for i in range(len(metadata)):
    abstracts[metadata['paper_title'].iloc[i]] = (
        model_sentences[i] + "" +
        metric_name_sentences[i] + "" +
        dataset_sentences[i] + "" +
        metric_value_sentences[i] + "" +
        lr_sentences[i] + "" + 
        bs_sentences[i] + "" + 
        mom_sentences[i] + "" +
        epoch_sentences[i] + "" +
        framework_sentences[i]
    )
    
# abstracts
```

### dataframe generation from list object

```python
abstract_df = pd.DataFrame(abstracts.items(), columns=['paper_title', 'abs'])

abstract_df.head()
```
